## **Step 2: VaR and CVaR Computation**
- **Objective**: Calculate VaR and CVaR using three distinct methods to understand potential downside risks.

**Procedure**:
1. **Historical Method**: Determine the 5th percentile (for a 95% VaR) directly from the historical return distribution.

2. **Variance-Covariance Method**: Assume returns are normally distributed and use the mean and standard deviation of returns to compute VaR.

3. **Monte Carlo Simulation**: Simulate potential future stock prices using Geometric Brownian Motion and derive the VaR from the simulated return distribution.


----

##### 1. Historical Method: 
- **Objective**: Determine the 5th percentile (for a 95% VaR) directly from the historical return distribution.

**Procedure**:

1. Calculate the 21 day rolling day returns - % 
2. Compute Var and CVar using those returns
3. Visualization


```python
### 1. Calculate the 21 day rolling day returns

def returns_days(data,DAYS):
    # Calculating rolling 30-trading-days returns
    data['Price x days ago'] = data['Adj Close'].shift(DAYS)
    data['Rolling x-day Return'] = (data['Adj Close'] - data['Price x days ago']) / data['Price x days ago']

    # Dropping the first 30 days since they won't have 30-day returns
    data_x_day_returns = data['Rolling x-day Return'].dropna()

    return data_x_day_returns


data_returns_tf = returns_days(data,DAYS)

data_returns_tf.head()
```

```python
## 2. Compute Var and CVar using those returns

def historical_V_CV(data_30_day_returns,confidence):
    
#     historical_var = -np.percentile(daily_returns, confidence * 100)
    historical_var = -data_30_day_returns.quantile(confidence)
    historical_cvar = -data_30_day_returns[data_30_day_returns < -historical_var].mean()
    
    return historical_var, historical_cvar #,data_30_day_returns

# Percentage values of VaR and CVaR 
historical_VaR_pc, historical_CVaR_pc =historical_V_CV(data_returns_tf,confidence) 

# Calculation $ values of VaR and CVaR
historical_VaR = InitialInvestment*historical_VaR_pc
historical_CVaR =InitialInvestment*historical_CVaR_pc

print("Historical Method")
print('Value at Risk 95th CI    :      ', round(historical_VaR,2))
print('Conditional VaR 95th CI  :      ', round(historical_CVaR,2))



```

```python
# Graph to Visualization

import seaborn as sns
 
plt.figure(figsize=(14, 7))

plt.axvline(-historical_VaR_pc,color='red',linewidth=3, label='VAR_Historical')
plt.axvline(-historical_CVaR_pc,color='orange',linewidth=2, label='CVAR_Historical')
sns.histplot(data=data_returns_tf.dropna(), kde=True, bins=100, color='blue')

plt.figtext(0.2, 0.7, "Mean : {} \nStd: {} ".format(data_returns_tf.mean(),data_returns_tf.std()))
plt.legend()
plt.grid(True)
plt.title('S&P 500 Historical '+str(DAYS) + ' Day Returns')
plt.show()
```

##### 1. Variance-Covariance Method : 
- **Objective**: Determine the 5th percentile (for a 95% VaR) by assuming returns are normally distributed and use the mean and standard deviation of returns to compute VaR and CVaR.

**Procedure**:

1. Develope function to calculate VaR and CVaR using VAR_COV method - use equations to do so
2. Compute Var and CVar using that function

---
- Formula for VaR and CVaR for normal distribution

```python
# Computing VaR using the Variance Covariance approach

from scipy.stats import norm  # importing normal distribution


def COV_VAR_V_CV(confidence,mean,std):
    
    # variance_covariance_V =  -(mean - std * norm.ppf((1-confidence)))
    # variance_covariance_V =  -(mean + std * norm.ppf(confidence))
    variance_covariance_V =  -norm.ppf(confidence, mean, std)

    variance_covariance_CV = -( mean -  std / confidence * norm.pdf (norm.ppf(confidence)) )
      
    return variance_covariance_V,variance_covariance_CV

# Percentage values of VaR and CVaR 
var_cov_VaR_pc,var_cov_CVaR_pc = COV_VAR_V_CV(confidence,data_returns_tf.mean(), data_returns_tf.std())#*data['Adj Close'].iloc[-1]

# Calculation $ values of VaR and CVaR
var_cov_VaR =InitialInvestment*var_cov_VaR_pc
var_cov_CVaR  =InitialInvestment*var_cov_CVaR_pc


print("Variance-Covariance Method")
print('Value at Risk 95th CI    :      ', round(var_cov_VaR,2))
print('Conditional VaR 95th CI  :      ', round(var_cov_CVaR,2))
```

##### 1.  **Monte Carlo Simulation**: 
- **Objective**: Simulate potential future stock prices using Geometric Brownian Motion and derive the VaR from the simulated return distribution.

**Procedure**:

1. Develope the simulation using Geometric Brownian Motion to forecast daily returns of the stock
2. Do multiple simulations to forecast the price after 21 days and visualize it.
3. Compute Var and CVar using the obtained simulated price after 21 days

------

- ##### Geometric Brownian Motion (GBM)

The formula to simulate stock prices using GBM is:

$$
S_t = S_0 \times \exp \left( \left( \mu - \frac{\sigma^2}{2} \right) \times t + \sigma \times W_t \right)
$$

Where:
- $( S_t )$ = Stock price at time $ t $
- $( S_0 $) = Initial stock price (at \( t = 0 \))
- $( \mu $) = Drift (mean return of the stock)
- $( \sigma $) = Volatility (standard deviation of the stock's returns)
- $( t $) = Time increment
- $( W_t $) = Random component (Brownian motion), which is a random sample from a standard normal distribution

The term $( \left( \mu - \frac{\sigma^2}{2} \right) \times t $) represents the deterministic trend (drift), and $( \sigma \times W_t $) is the stochastic term (random shock) that introduces volatility into the model.

---


```python
# Create a function for GBM simulation 
import numpy as np 

def simulate_gbm(S0, mu,sigma, days, num_simulations):
    
    # Time increment (daily)
    ## As we have daily returns - instead of scaling time,mu,sigma statistically, 
    ## Lets just make this simple by always having increament of one day.             
    dt = 1 
    
    ## Creating simulation array of number of days * num of simulations 
    simulations = np.zeros((days+1, num_simulations))
    
    ## # Calculating Random component assuming normal distribution
    wt_array = np.random.normal(size = (days,num_simulations))
    
    for i in range(num_simulations):  # for every simulation
        
        price_path = [S0] # initial price
                
        for j in range(days): # for each day 
            
                
            Wt = wt_array[j,i]  
            price = price_path[-1] * np.exp((mu - 0.5 * sigma**2) * dt + sigma * Wt) ## Check GBM 
            price_path.append(price)
            
            
        simulations[:, i] = price_path[:]
        
    return simulations


## Simulation Parameters 
S0 = data['Adj Close'].iloc[-1]
mu =  meanReturns #meanReturns = data['Daily Return'].mean()
sigma = stdReturns #stdReturns = data['Daily Return'].std()
num_sim = 10000 # number of simulations
num_days = DAYS # Predicting price after how many days - steps


GBM_Simulation = simulate_gbm(S0, mu,sigma, num_days, num_sim)

## Plot Simulation results
plt.figure(figsize=(14, 7))
plt.plot(GBM_Simulation)
plt.xlabel("Trading Days")
plt.ylabel("Stock Price")
plt.title(
    " Realizations of Geometric Brownian Motion \n $dS_t = \mu S_t dt + \sigma S_t dW_t$\n $S_0 = {0}, \mu = {1} \%, \sigma = {2} \% $\n Number of Simulations = {3} ".format(round(S0,2), round(mu,6), round(sigma,6),num_sim)
)
plt.grid(True)
plt.tight_layout()
plt.xticks(rotation=45)  
plt.show()
```

```python
# Simulated price after 21 days for all simulation

Final_sim_price = GBM_Simulation[DAYS,:]

# Plot the distribution
plt.figure(figsize=(14, 7))

sns.histplot(data=Final_sim_price, kde=True, bins=100, color='blue')
plt.axvline(Final_sim_price.mean(),color='red',linewidth=3, label='mean')
plt.title('Monte Carlo Simulation')
plt.figtext(0.2, 0.7, "Mean_Price : {} \nInitial Price : {}".format(Final_sim_price.mean(),S0))
plt.legend()
plt.xlabel('S&P 500 GBM Simulated price after '+str(DAYS) + ' Days')

plt.grid(True)
plt.show()
```

3. Compute Var and CVar using the obtained simulated price after 21 days

```python
# Calculatio Var and CVar after 21 days

mc_VaR_price = np.percentile(Final_sim_price, confidence * 100) 
mc_CVaR_price = (Final_sim_price[Final_sim_price < mc_VaR_price]).mean() 

mc_VaR_pc =(S0 - mc_VaR_price)/S0
mc_CVaR_pc = (S0 - mc_CVaR_price)/S0

mc_VaR = mc_VaR_pc  * InitialInvestment
mc_CVaR = mc_CVaR_pc  * InitialInvestment

print('Value at Risk 95th CI    :      ', round(mc_VaR,2))
print('Conditional VaR 95th CI  :      ', round(mc_CVaR,2))
```

-----

Stress testing is an essential tool in risk management that allows you to understand the potential impact of adverse market scenarios on your portfolio. Unlike VaR and CVaR, which provide potential loss values based on historical data or model assumptions, stress testing is explicitly designed to evaluate the impact of extreme but plausible adverse conditions.


----
